```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Copy of Dataset for Data Analytics.xlsx - Sheet1.csv")
```

```python
df.describe()
```

```python
print("Mean Total Price:", df["TotalPrice"].mean())
print("Median Total Price:", df["TotalPrice"].median())
print("Total Orders:", df["OrderID"].count())
```

```python
q1 = df["TotalPrice"].quantile(0.25)
q3 = df["TotalPrice"].quantile(0.75)
iqr = q3 - q1

outliers = df[(df["TotalPrice"] < (q1 - 1.5 * iqr)) | (df["TotalPrice"] > (q3 + 1.5 * iqr))]
print(outliers)
```

```python
print(df["OrderStatus"].value_counts())
print(df["PaymentMethod"].value_counts())
print(df["OrderStatus"].value_counts())
```

```python
df["PaymentMethod"].value_counts().plot(kind='bar')
plt.title("Payment Method")
plt.show()
```

```python
df["TotalPrice"].plot(kind="hist", bins=20)
plt.title("Total Price Distribution")
plt.show()
```
